```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
import time

# Список текстов для перевода
texts = [
    "Уважаемый сэр.\nЯ еще не подтвердил своевременное получение\nвашего письма от 31-го числа прошлого месяца\n<<1>>и хочу поблагодарить вас за те сведения, которые\nвы мне сообщили в ответ на мои вопросы о России.\nНадеюсь, что впредь этот район более не даст нам\nповода для тревоги, и хотя я считаю, что\nбезрассудство покойного императора и\nпоследствия этого весьма затруднили мирное\nразвитие отношений между Англией и Россией, все\nже есть основания надеяться, что этого можно\nдостигнуть. Я называю это безрассудством покойного\nимператора, ибо считаю, что нельзя употребить\nболее мягкого выражения. Эта северная война —\nрезультат его прихоти, а тысячи храбрых датчан,\nпогибших при обороне своей столицы, — кровавая\nжертва, принесенная его призраку. Храни нас\nГосподь от повторения подобных гекатомб.\nГоворя это, я должен признаться, что не\nпредставляю себе, как новый император сможет\nотказаться от основного принципа вооруженного\nнейтралитета или как без отказа от него можно\nизбежать войны.\nИбо\nболее невозможно\nзапугать английское\nправительство, так же как и убедить его\nотказаться от своей позиции в этом вопросе...\n<<2>>\nВаш...\nA.\nПолучено 16 апреля, ответ дан 21 апреля\n<<3>>.",
    "Излагает мнение русского дипломата А. И. Муравьева о благожелательном отношении его правительства к торговле с США."
]

# Настройки браузера
chrome_options = Options()
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36")

# Путь к ChromeDriver
service = Service(executable_path='/path/to/chromedriver')  # Замените на реальный путь
driver = webdriver.Chrome(service=service, options=chrome_options)

try:
    driver.get("https://slavenica.com/ ")

    # Явное ожидание загрузки поля ввода
    text_input = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.ID, "text_input"))
    )

    for i, text in enumerate(texts):
        # Очистка и ввод текста
        text_input.clear()
        text_input.send_keys(text)
        print(f"Текст {i+1} введен")

        # Выбор исходного языка (например, "Русский" → "Старославянский")
        source_lang = driver.find_element(By.ID, "source_language")
        source_lang.click()
        driver.find_element(By.XPATH, "//select[@id='source_language']/option[text()='Русский']").click()

        # Выбор целевого языка (например, "Старославянский" → "Русский")
        target_lang = driver.find_element(By.ID, "target_language")
        target_lang.click()
        driver.find_element(By.XPATH, "//select[@id='target_language']/option[text()='Старославянский']").click()

        # Нажатие кнопки перевода
        translate_button = driver.find_element(By.ID, "translate_button")
        translate_button.click()
        print("Кнопка перевода нажата")

        # Ожидание результата
        result = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.ID, "translation_result"))
        )
        print(f"\nОригинал:\n{text}\n\nПеревод:\n{result.text}\n{'-'*50}")

        time.sleep(3)  # Пауза между запросами

except Exception as e:
    print("Ошибка:", str(e))
    driver.save_screenshot("error.png")

finally:
    driver.quit()
```
